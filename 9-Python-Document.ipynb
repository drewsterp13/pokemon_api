```python
def get_racer_info(data):
    new_data = []
    for racer in data:
        racer_dict = {}
        racer_name = f'{racer["Driver"]["givenName"]} {racer["Driver"]["familyName"]}'
        
        racer_dict[racer_name] = {
            "first_name": racer["Driver"]["givenName"],
            "last_name": racer["Driver"]["familyName"],
            "position": racer["position"],
            "wins": racer["wins"],
            "DOB": racer["Driver"]["dateOfBirth"],
            "nationality": racer["Driver"]["nationality"],
            "constructor": racer["Constructors"][0]["name"]
        }
        new_data.append(racer_dict)
    return new_data

get_racer_info(my_racer_data)
```

```python
def get_driver_info_by_year_and_round(year, r):
    url = f"http://ergast.com/api/f1/{year}/{r}/driverStandings.json"
    response = requests.get(url)
    if not response.json()["MRData"]["StandingsTable"]["StandingsLists"]:
        return "We had an error loading your data :( It is likely the year or round is not in the database"
    
    data = response.json()["MRData"]["StandingsTable"]["StandingsLists"][0]["DriverStandings"]
    return get_racer_info(data)

get_driver_info_by_year_and_round(2019, 4)
```

Use the requests package to connect to this API and get and store data for 20 different pokemon.
Get the pokemons: name, atleast one ability's name, base_experience, and the URL for its sprite (an image that shows up on screen) for the 'front_shiny', attack base_state, hp base_stat, defense base_stat

```python
import requests, json
import random

# With this program, there is a list of ids that when put into a function, it gets the pokemon info and stores it
# inside a dictionary.

list_url = "https://pokeapi.co/api/v2/pokemon?limit=50&offset=0" # This is the library url
get_list = requests.get(list_url) # This gets the json into the memory as a dictonary
pokechives = {} # This is the local dictionary where the pokemon info will be stored

def getPokemonInfo(pokelist): # This goes thru a list of pokedex numbers
    index = 0 # This is the index for the pokemon list
    
    print(f"You have {len(pokelist)} pokemon that you have catched")
    print()
    
    for ID in pokelist:
        
        get_pokemon = get_list.json()["results"][ID-1]["url"] # Gets the pokemon info using the id from the pokelist
        get_stats = requests.get(get_pokemon) # Gets the json from the info
        name = get_stats.json()["name"] # This gets the name
        base_experience = get_stats.json()["base_experience"] # This gets the base experience
        front_shiny = get_stats.json()["sprites"]["front_shiny"] # This gets the sprite for the pokemon
        
        pokechives[index] = {
            "name": name,
            "id": ID,
            "base_experience": base_experience,
            "base_stats": {},
            "types": [],
            "abilities": [],
            "sprite": front_shiny
        }
        
        for i in range(3): # This goes thru the stats
            stat = get_stats.json()['stats'][i]['stat']['name']
            number = get_stats.json()['stats'][i]['base_stat']
            pokechives[index]["base_stats"][stat] = number
        for i in range(len(get_stats.json()["abilities"])): # This goes thru the abilities
            ability = get_stats.json()["abilities"][i]['ability']['name']
            pokechives[index]["abilities"].append(ability)
        for i in range(len(get_stats.json()["types"])): # This goes thru the types
            poketype = get_stats.json()["types"][i]["type"]["name"]
            pokechives[index]["types"].append(poketype)
        
        # This displays the pokemon info
        for key, value in pokechives[index].items():
            print(f"{key}: {value}")
        print()
        
        index += 1
    
    print("We hope that you enjoyed the basic info about the pokemon that you catched")
    
    
# Enter HERE the pokedex numbers to learn about the pokemon that you selected
pokemon_list = [25, 7, 39, 13, 6, 2, 23, 18, 21, 35, 12, 10, 1, 8, 13, 26, 46, 16, 28, 9] #(SELECTED POKEMON)

# This randomly selects pokemon from pokedex #0001 to #0050
# pokemon_list = []
# for i in range(20):
#     i = random.randint(1, 50)
#     pokemon_list.append(i)

# This activates the function
getPokemonInfo(pokemon_list)
```
